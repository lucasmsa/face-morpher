## Morfar faces 
- Recebendo duas imagens de input, gerar um output morfando ambas 
- A parte de morfar ser√° realizada frame por frame, em que cada frame entre t (variando de 0 a 1), cada pixel (x, y) ter√° o frame morfado atrav√©s da seguinte f√≥rmula $M_t(x, y) = (1 - t)*I_0(x, y) + t*I_1(x, y)$; logo, quando t for igual a 0, a imagem morfada √© igual a $I_0$, se for 1, a imagem morfada √© igual a $I_1$, atrav√©s de um m√©todo chamado _cross-dissolve_ que balan√ßa a colora√ß√£o nos frames do meio, mas deixar√° um efeito de _ghosting_ visto que as faces n√£o est√£o alinhadas
- A seguir, uma demonstra√ß√£o de como ficariam duas imagens morfadas caso apenas seja usada essa f√≥rmula

### Adicionando uma fun√ß√£o que vai extrair a face de uma pessoa dada uma determinada foto, recebendo, os pixels da imagem e uma normaliza√ß√£o que deixar√° a imagem em um tamanho fixado. Se uma face for detectada, a face normalizada ser√° retornada, caso contr√°rio, a imagem completa normalizada √© retornada ü¶Å

```python
def extract_faces(img_pixels, normalization_scale):
    face_detection = cv2.CascadeClassifier('./utils/haarcascade_frontalface_default.xml')
    img_pixels_grayscale = cv2.cvtColor(img_pixels, cv2.COLOR_BGR2GRAY) 
    face_pixels = face_detection.detectMultiScale(img_pixels_grayscale, scaleFactor=1.1, minNeighbors=3, minSize=(20, 20))
    try:
        face = img_pixels[face_pixels[0][1]:face_pixels[0][1] + face_pixels[0][3], \
                            face_pixels[0][0]:face_pixels[0][0] + face_pixels[0][2]]
    except:
        face = img_pixels
    finally:
        return cv2.resize(face, normalization_scale)
```

### Testar o _morph_ com uma simples fun√ß√£o que deixar√° um efeito de ghosting, essa fun√ß√£o vai fazer a m√©dia ponderada para cada ponto da imagem, com um piv√¥ _t_, esse est√° no intervalo \[0, 1\]. Para _t_ maior que 0.5, a imagem predominante ser√° a segunda, caso _t_ menor que 0.5 a primeira prevalece, caso contr√°rio um mistura homog√™nea de ambas ser√° realizada ü•¥

```python
def face_morpher_formula_ghosting(t, point_img_1, point_img_2):
    return (1 - t)*point_img_1 + t*point_img_2

def face_morpher_ghosting(t, img_1, img_2):
    size = img_1.shape
    morphed_img = img_1.copy()

    for height in range(size[0]):
        for width in range(size[1]):
            point_img_1 = img_1[height][width]
            point_img_2 = img_2[height][width]
            morphed_img[height][width] = [round(face_morpher_formula_ghosting(t, point_img_1[rgb], point_img_2[rgb])) \
                                            for rgb in range(3)]
    return morphed_img

face_1, face_2 = extract_faces(img_1, (256,256)), extract_faces(img_2, (256,256))
morph_with_ghosting = face_morpher_ghosting(0.5, face_1, face_2)
show_multiple_images([face_1, morph_with_ghosting, face_2], 1, 3, 'Faces morph with ghosting')
```

### Visando corrigir a problem√°tica, faz-se necess√°rio corrigir a parte geom√©trica das imagens, o efeito de ghosting ocorre porque o alinhamento entre os elementos faciais das pessoas da foto n√£o est√° ideal. Para alinhar as formas entre o input (duas imagens) e o output (imagem morfada) estejam sendo bem correspondidas nos frames pr√©-output. Para cada par de ponto cedido, calcula-se uma nova localiza√ß√£o do ponto correspondente em cada frame morfado para criar uma m√©dia do formato deformado da imagem

```python
def paint_face_landmarks(img_1, face_landmarks):
    painted_image = img_1.copy()
    width, height, _ = img_1.shape
    for key, landmark in face_landmarks.items():
        for pixel in landmark:
            for i in range(-15, 15):
                if pixel[0] < width and pixel[1] < height: 
                    painted_image[pixel[0] + i, pixel[1] + i] = (255, 0, 0)

    show_image(painted_image)

face_landmarks_list = face_recognition.face_landmarks(img_1)
print(face_landmarks_list)
paint_face_landmarks(img_1, face_landmarks_list[0])
```

```python
def extract_face_landmarks(img, detector, predictor, correspondance, size, filename):
    landmarks_list = []
    bounding_box_detection = detector(img, 1)

    try:
        if not bounding_box_detection:
            raise NoFaceFound
    except NoFaceFound:
        print('No face could be found in the image')
    
    for idx, rectangle in enumerate(bounding_box_detection):
        
        shape = predictor(img, bounding_box_detection[0])

        for i in range(68):
            x = shape.part(i).x
            y = shape.part(i).y
            print(x,y, shape.part(i))
            if x >= size[0]: x = size[0] - 1
            if y >= size[1]: y = size[1] - 1
            landmarks_list.append((x, y))
            correspondance[i][0] += x
            correspondance[i][1] += y

        landmarks_list.append((1,1))
        landmarks_list.append((size[1]-1,1))
        landmarks_list.append(((size[1]-1)//2,1))
        landmarks_list.append((1,size[0]-1))
        landmarks_list.append((1,(size[0]-1)//2))
        landmarks_list.append(((size[1]-1)//2,size[0]-1))
        landmarks_list.append((size[1]-1,size[0]-1))
        landmarks_list.append(((size[1]-1),(size[0]-1)//2))

    cv2.imwrite(f'./output/{filename}.png', img)

    return landmarks_list
```

### Nessa fun√ß√£o, os principais pontos de ambas as faces seram demarcados (nesse caso, um algoritmo vai selecionar 68 pontos faciais das imagens) para cada uma das imagens sem alterar o _background_ dessa. Destarte, a correspond√™ncia dos pontos √© entrela√ßada entre as imagens e o mesmo processo de manuten√ß√£o do _background_ √© feito novamente

```python
def get_faces_main_points(img_1, img_2):
    detector = dlib.get_frontal_face_detector()
    predictor = dlib.shape_predictor('./utils/shape_predictor_68_face_landmarks.dat')
    correspondances = np.zeros((68, 2))

    size = (img_1.shape[0], img_2.shape[1])
    first_image_landmarks = extract_face_landmarks(img_1, detector, predictor, correspondances, size, 'img_1')
    second_image_landmarks = extract_face_landmarks(img_2, detector, predictor, correspondances, size, 'img_2')

    narray = correspondances/2
    narray = np.append(narray,[[1,1]],axis=0)
    narray = np.append(narray,[[size[1]-1,1]],axis=0)
    narray = np.append(narray,[[(size[1]-1)//2,1]],axis=0)
    narray = np.append(narray,[[1,size[0]-1]],axis=0)
    narray = np.append(narray,[[1,(size[0]-1)//2]],axis=0)
    narray = np.append(narray,[[(size[1]-1)//2,size[0]-1]],axis=0)
    narray = np.append(narray,[[size[1]-1,size[0]-1]],axis=0)
    narray = np.append(narray,[[(size[1]-1),(size[0]-1)//2]],axis=0)
    
    return {
        "size": size,
        "first": first_image_landmarks,
        "second": second_image_landmarks,
        "correspondances": narray
    }

correspondance_points = get_faces_main_points(face_1, face_2)
```
